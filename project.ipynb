Sales Prediction using Python

```python
# Loading the important packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
# Loading the data set 
df=pd.read_csv(r"C:\Users\jsuya\Documents\Sales Prediction using Python\advertising.csv")

# checking the data  is clean or not 
df.isnull().sum() # data is clean 

# Data Visullization & Exploration
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Features Correlations")
plt.show()

# split the data (traning and testing purpose)
x=df[['TV','Radio','Newspaper']]
y=df['Sales']
x_trian,x_test,y_trian,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

# Trian the model
model=LinearRegression()
model.fit(x,y)

# make predications 
y_pred=model.predict(x_test)

# Evaluate the Model
print("============================================================================")
print("R_2 Score : ",r2_score(y_test,y_pred))
print("mean_absolute_error : ",mean_absolute_error(y_test,y_pred))
print("mean_squrt_error :  ",mean_squared_error(y_test,y_pred))
print("Root_mean_sqrt_error",np.sqrt(mean_squared_error(y_test,y_pred)))
print("=================================================================================")

# comparing the actual and predicated data 
compare=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
compare.head()

# visualize
plt.scatter(y_test,y_pred,color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("Actual Sales")
plt.ylabel("Predicated Sales")
plt.title("Actual vs Predicated Sales")
plt.show()

# test on this Real Data 
tv=float(input(" Enter the tv advertisement "))
Radio=float(input(" Enter the  Radio advertisement "))
Newspaper=float(input(" Enter the Newspaper advertisement "))
predication=model.predict([[tv,Radio,Newspaper]])
print("predicated sales for givean data ",predication.astype(int))


```
